# Уровни общения
- [уровни общения](https://www.youtube.com/watch?v=A_o7Pg3TA5o&t=39s)
    - С друзьями, родными, супругами
    - душа другого человека, потемки
        - там растут деревья, кусты. Мы это видим, как общую картину. А что творится там, в этих кустах, в этой роще? Мы не знаем этого.
        - другой человек также спрятан в своей роще. Так как они общаются? Как выйти из этих зарослей, из лесу.
        - при общении люди подходят к самому краюшку своих лесных зарослей. И они видят его на фоне среды. Но они не знают о чем этот человек думает, о чем мечтает. Ничего не знают.
        - Давайте мы представим, на какое расстояние человек может приблизится, в жизнь другого человека.
    - Человек Существо общественное, он не может быть сам.
    - 1й уровень
        - клише
    - 2й уровень - факты
        - деловые встречи;
        - сообщения;
        - нет близких отношений;
        - (никаких секретов, тайн)
    - 3й уровень - идеи
        - требует времени и внимания (язык времени)
        - уровень идей (о том, что я хотел бы, уровень желания)
        - не раскрывает эмоций
        - отношения дружественные
    - 4й уровень - разговаривают, передавая идеи друг другу
    - на этом уровне мы хорошо узнаем человека (открываем свой эмоциональный мир)
        - уровень чувств
        - высказыване негативных эмоций
        - беседы на интеллектуальном уровне
        - доверительный тон
        - умение слушать
        - эмоциональные реакции
    - 5й уровень - самораскрытие
    - семейный союз
        - полное доверие
        - нет тайн друг от друга
        - семья - единое целое
        - потрясения не разрушает семью
