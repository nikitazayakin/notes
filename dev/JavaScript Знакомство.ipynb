# Switch

Конструкция switch заменяет собой сразу несколько if

switch (a) {
    case 3:
        console.log ( "three" );
        break;
    case 4:
        console.log ( "four" );
        break;
    case 5:
        
    case undefined:
        console.log ( " undefined " );
        break;

        console.log ( "some other value" );
}
